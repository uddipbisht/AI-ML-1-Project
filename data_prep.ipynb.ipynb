```python
def team_summary(data):
    noc, year = data.name
    return pd.Series({
        'team': noc,
        'country': data["Team"].mode().iloc[0] if not data["Team"].mode().empty else None,
        'year': year,
        'events': data['Event'].nunique(),
        'athletes': len(data),
        'age': data["Age"].mean(skipna=True),
        'height': data["Height"].mean(skipna=True),
        'weight': data["Weight"].mean(skipna=True),
        'medals': data["Medal"].notna().sum()
    })
```

```python
def prev_medals(data):
    data = data.sort_values("year", ascending=True)
    data["prev_medals"] = data["medals"].shift(1)
    data["prev_3_medals"] = data["medals"].rolling(3, closed="left", min_periods=1).mean()
    return data
```
